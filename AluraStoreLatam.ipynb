## 1. Introducción

### 1.1 Contexto del análisis

El presente informe tiene como objetivo analizar el desempeño de las cuatro tiendas que conforman la cadena **Alura Store**, con el fin de identificar cuál de ellas presenta el menor rendimiento comercial y operativo.

El Sr. Juan necesita esta información para tomar una decisión estratégica sobre **qué tienda vender**, con el objetivo de obtener capital que le permita invertir en un nuevo negocio.

Para apoyar esta decisión, se realizó un análisis de datos basado en distintos indicadores clave de desempeño de cada tienda.

---

### 1.2 Metodología

El análisis se realizó utilizando datos históricos de ventas y evaluando cinco aspectos principales:

- **Ingresos totales por tienda**
- **Categorías de productos más y menos vendidas**
- **Calificaciones promedio de los clientes**
- **Productos más y menos vendidos**
- **Costo promedio de envío**

A partir de estos datos se generaron diferentes visualizaciones con **Matplotlib**, las cuales permiten comparar el desempeño de cada tienda e identificar patrones relevantes en las ventas.
